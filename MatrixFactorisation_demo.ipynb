```python
from collections import defaultdict

def get_top_n(predictions, n=10):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
    
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)  # Sort by estimated rating
        top_n[uid] = user_ratings[:n]  # Keep top-N

    return top_n

top_n_predictions = get_top_n(predictions, n=5)
print(top_n_predictions)
```

```python
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split
import pandas as pd
import numpy as np
from scipy.sparse import coo_matrix

# ratings_df = Dataset.load_from_df(data[['UserId','MovieId','Rating']], reader)

# Define the dataset for Surprise
reader = Reader(rating_scale=(1, 5))
dataset = Dataset.load_from_df(data[["UserId", "MovieId", "Rating"]], reader)

# Train-test split
trainset, testset = train_test_split(dataset, test_size=0.2)

# Train the SVD model
model = SVD()
model.fit(trainset)

# Create user-item matrix
user_item_matrix = data.pivot(index="UserId", columns="MovieId", values="Rating")

# Convert user-item matrix to a sparse matrix (CSR format)
user_item_matrix_sparse = coo_matrix(user_item_matrix.fillna(0))

# Extract missing values from the sparse matrix
missing_values = np.where(user_item_matrix_sparse.toarray() == 0)
# Ensure that the testset is correctly structured: a list of (user, item, rating) tuples
all_user_item_pairs = [
    (uid, iid, 0) for uid in data["UserId"].unique() for iid in data["MovieId"].unique()
]

# You need to predict on missing values only
predictions = model.test(all_user_item_pairs)


# # Predict ratings for missing values
# predictions = model.test(zip(missing_values[0] + 1, missing_values[1] + 1))  # +1 because Surprise uses 1-based indexing

# Map predictions back to the user-item matrix
predicted_ratings = { (pred.uid, pred.iid): pred.est for pred in predictions }

# Update the original user-item matrix with predicted values for missing ratings
for i, j in zip(missing_values[0], missing_values[1]):
    if (i + 1, j + 1) in predicted_ratings:
        user_item_matrix.iloc[i, j] = round(predicted_ratings[(i + 1, j + 1)], 2)

print(user_item_matrix)
```

```python
# Loop through missing values and replace them with predicted ratings
for i, j in zip(missing_values[0], missing_values[1]):
    # Get the user and movie IDs (we need to get these from the matrix indices, not integer indices)
    user_id = user_item_matrix.index[i]
    movie_id = user_item_matrix.columns[j]

    # Check if the prediction for this (user, movie) pair exists
    if (user_id, movie_id) in predicted_ratings:
        # Replace the NaN value with the predicted rating
        user_item_matrix.loc[user_id, movie_id] = round(predicted_ratings[(user_id, movie_id)], 2)

print(user_item_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(user_item_matrix)
user_similarity
```
